```python
emmatokens = nltk.word_tokenize(text)
```

```python
emmatokens = nltk.word_tokenize(text)

emmawords = [w.lower( ) for w in emmatokens] 
```

```python
# Creating a frequency distribution of words
ndist = FreqDist(emmawords)
```

```python
# print the top 30 tokens by frequency
nitems = ndist.most_common(5)
for item in nitems:
    print (item[0], '\t', item[1])
```

```python
# Bigrams and Bigram frequency distribution
emmabigrams = list(nltk.bigrams(emmawords))
print(emmawords[:5])
print(emmabigrams[:5])
```

```python
ndistbi = FreqDist(emmabigrams)
```

```python
type(ndistbi)
```

```python
# print the top 30 tokens by frequency
ndistL = []
for k,v in ndistbi.items():
    ndistL.append([k[0],k[1],v])
```

```python
import pandas as pd
ndist_df = pd.DataFrame(ndistL, columns=['w1', 'w2', 'freq'])
```
